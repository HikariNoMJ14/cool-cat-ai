```python
import os
import sys
import glob
import pandas as pd
from matplotlib import rcParams

rcParams['figure.figsize'] = 20,10

sys.path.append('..')
```

```python
filepath = '../data/encoded/duration/mono/*/*.csv'

all_encoded = glob.glob(filepath)
```

```python
all_durations = set({})
for f in all_encoded:
    e = pd.read_csv(f, index_col=0)
    dur = set(e['duration'].value_counts().index.astype('int').values)

    all_durations.update(dur)
    print(len(all_durations))
```

```python
all_durations
```

```python
all_df = []
for f in all_encoded:
    e = pd.read_csv(f, index_col=0)
    all_df.append(e['duration'])
    
all_df = pd.concat(all_df)
```

```python
all_df.value_counts().tail(30)
```

```python
import seaborn as sns

sns.countplot(y=0, data=all_df, orient='v')
```
