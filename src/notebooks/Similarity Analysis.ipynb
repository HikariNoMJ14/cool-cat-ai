```python
from tslearn.metrics import dtw
import numpy as np
import pandas as pd
import sys
import itertools
import seaborn as sns
import matplotlib.pyplot as plt

sys.path.append('../..')

from src.utils import get_filepaths
```

```python
original_filepaths = get_filepaths('original')
improvised_filepaths = get_filepaths('improvised')
```

```python
columns = ['pitch', 'duration']
```

```python
sims = []

# for f1, f2 in list(itertools.combinations(improvised_filepaths, 2)):

for f1 in improvised_filepaths:
    print(f1)
    m1 = pd.read_csv(f1, index_col=0)[columns]

    for f2 in improvised_filepaths:
        m2 = pd.read_csv(f2, index_col=0)[columns]

        sim = dtw(m1, m2)

        sims.append({
            'm1': "/".join(f1.split('/')[-2:]),
            'm2': "/".join(f2.split('/')[-2:]),
            'sim': sim
        })

sims_df = pd.DataFrame.from_dict(sims)
sims_df['norm_sim'] = sims_df.sim.apply(lambda x: (x-sims_df.sim.min())/(sims_df.sim.max()-sims_df.sim.min()))
```

```python
sns.distplot(sims_df['norm_sim'])
```
