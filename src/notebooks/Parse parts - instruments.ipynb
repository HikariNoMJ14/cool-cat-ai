```python
from mido import MidiFile
```

```python
mid = MidiFile(file, clip=True)
```

```python
mid.tracks
```

```python
track1 = mid.tracks[0]
```

```python
notes = {}

for t in mid.tracks:
    for m in t:
#         print(m)
        if m.type == 'note_on':
            if m.channel not in notes:
                notes[m.channel] = []
            notes[m.channel].append(m)
        
for c, n in notes.items():
    print(len(n))
```
