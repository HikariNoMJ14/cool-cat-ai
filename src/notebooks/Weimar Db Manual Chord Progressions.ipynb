```python
import pandas as pd
import json
import numpy as np
```

```python
df = pd.read_csv('../data/sources/WeimarDB/wjazzd_manual_chords.csv')
```

```python
df
```

```python
margie = df[df['melid'] == 27].reset_index()
margie
```

```python
oleo = df[df['melid'] == 231].reset_index()
oleo
```

```python
blue = df[df['melid'] == 218].reset_index()
blue
```

```python
margie['step'] = (margie['bar'] - 1) * 4 + (margie['beat'] - 1)
oleo['step'] = (oleo['bar'] - 1) * 4 + (oleo['beat'] - 1)
blue['step'] = (blue['bar'] - 1) * 4 + (blue['beat'] - 1)
```

```python
margie[['bar', 'beat', 'step', 'chord']]
```

```python
blue[['bar', 'beat', 'step', 'chord']]
```

```python
oleo[['bar', 'beat', 'step', 'chord']]
```

```python
sections = []
parts = {}

current_part = None
current_chord = None

for i, r in margie.iterrows():
    form = r['form']
    chord = r['chord']
    
    if type(form) == str:
        sections.append(form)
        current_part = form
        
        parts[form] = []
    
    if type(chord) == str:
        current_chord = chord
        parts[current_part].append(chord)
    else:
        parts[current_part].append(current_chord)
    
    print(form, chord, current_part, current_chord)    
```
