```python
import numpy as np
import pandas as pd
import torch
import seaborn as sns
```

```python
dataset_c = torch.load(
'/media/manu/Data/PycharmProjects/thesis/data/tensor_dataset/duration/mono/2022_05_27_204824_transpose_c_chord_extended_7.pt'
)

li = []

for song in dataset_c:
    for row in song.tolist():
        di = {
            'offset': row[2],
            'pitch': row[3] if row[3] != 128 else np.nan,
            'duration': row[4],
            'chord': row[5:12]
        }
        li.append(di)
        
dfc = pd.DataFrame.from_dict(li)
dfc['pitch_class'] = dfc['pitch'].apply(lambda x: x % 12 if not np.isnan(x) else np.nan)
dfc['chord_pitch_class'] = dfc['chord'].apply(lambda x: [y % 12 if not np.isnan(y) else np.nan for y in x])
```

```python
sns.distplot(dfc['pitch_class'])
```

```python
chord_pitch_class = []

for idx, val in dfc['chord_pitch_class'].iteritems():
    chord_pitch_class += val

sns.distplot(chord_pitch_class)
```
