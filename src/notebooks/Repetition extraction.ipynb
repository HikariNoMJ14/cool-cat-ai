```python
folder = '../data/Complete Examples Melodies/Real Book'

filepaths = [y for x in os.walk(folder) for y in glob(os.path.join(x[0], '*.mid'))]

# filepaths = [
#     '../data/Complete Examples Melodies/Real Book/All Of Me.mid',
#     '../data/Complete Examples Melodies/Real Book/A Felicidade.mid',
#     '../data/Complete Examples Melodies/Real Book/Autumn Leaves.mid',
#     '../data/Complete Examples Melodies/Real Book/Fly Me To The Moon.mid',
#     '../data/Complete Examples Melodies/Real Book/How Insensitive.mid',
#         '../data/Complete Examples Melodies/Real Book/Meditation.mid',
#             '../data/Complete Examples Melodies/Real Book/One Note Samba.mid',
#             '../data/Complete Examples Melodies/Real Book/Summertime.mid',
#             '../data/Complete Examples Melodies/Real Book/Wave.mid',
#         '../data/Complete Examples Melodies/Doug McKenzie/Autumn Leaves.mid',
# ]

chord_progressions_filepath = '../data/chord_progressions/irb_chord_progressions.json'
chord_progressions = json.load(open(chord_progressions_filepath))
```

```python
m = Melody(filepaths[0])

m.setup()
m.set_song_structure(chord_progressions[m.song_name])
m.parse_notes()
m.align_melody()

measure = 0
bpm = m.time_signature[0]
tpb = m.mido_obj.ticks_per_beat
starting_measure = m.starting_measure

n_chord_prog_measures = int(np.sum([
    len(m.song_structure['progression'][section]) 
    for section in m.song_structure['sections']]
)  / m.time_signature[0])

n_measures = int(np.floor(np.max([
    x['measure'] + ((x['duration'] / tpb) / bpm) for x in m.note_info
])) + 1)


a = n_measures - starting_measure
o = a
r = 0

print(m.errors)
print(
    n_measures,
    np.max([x['measure'] for x in m.note_info]), 
    np.max([((x['duration'] / tpb) / bpm) for x in m.note_info])
)

while o > n_chord_prog_measures:
    r += 1
    o -= n_chord_prog_measures

print('intro', starting_measure)
print('total', n_measures)
print('cp', n_chord_prog_measures)
print('reps', r)
print('outro', o)

# print(len(m.note_info))
# print(len([n for n in m.note_info if starting_measure <= n['measure'] < n_measures - o]))

xxx = [n for n in m.note_info if starting_measure <= n['measure'] < starting_measure + n_chord_prog_measures]
```
