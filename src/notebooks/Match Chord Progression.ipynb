```python
import os
from glob import glob
import itertools

from utils import get_chord_progressions, filename_to_songname
from string_similarity import similar
```

```python
version = '1.2'
folder = f'../data/Complete Examples/v{version}'

files = [y for x in os.walk(folder) for y in glob(os.path.join(x[0], '*.mid'))]

data = [(l.split('/')[-2], l.split('/')[-1]) for l in files]
```

```python
chord_progressions = get_chord_progressions()
cp_names = set(chord_progressions.keys())
```

```python
cp_names
```

```python
s_names = set([filename_to_songname(n[1]) for n in data])
```

```python
s_names
```

```python
for cp, sn in list(itertools.product(cp_names, s_names)):
    score = similar(cp, sn)
    
    if score > 0.75 and score != 1:
        print(f'{cp} --- {sn}')
```
